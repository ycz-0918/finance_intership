```python
import pandas as pd
import os
import re

# 读取CSV文件
csv_path = "宏观（仅id）.csv"   # 改成你的文件路径
df = pd.read_csv(csv_path, encoding="gbk")

# 输出目录（可选）
output_dir = "output_宏观_pkl"
os.makedirs(output_dir, exist_ok=True)

# 清理文件名（避免非法字符）
def clean_filename(name):
    return re.sub(r'[\\/*?:"<>|]', "_", str(name))

# 按列保存为pkl
for col in df.columns:
    col_data = df[[col]]  # 保持DataFrame结构（不是Series）

    file_name = clean_filename(col) + ".pkl"
    file_path = os.path.join(output_dir, file_name)

    col_data.to_pickle(file_path)

print("完成：每一列已保存为独立pkl文件")
```

```python
import pandas as pd
import os

folder = "output_宏观_pkl"

for file in os.listdir(folder):
    if file.endswith(".pkl"):
        path = os.path.join(folder, file)
        df = pd.read_pickle(path)

        print(file, "行数:", len(df))
```

```python
import pandas as pd
import os
import re

csv_path = "指标（仅id）.csv"
df = pd.read_csv(csv_path, encoding="gbk")

output_dir = "output_pkl"
os.makedirs(output_dir, exist_ok=True)

def clean_filename(name):
    return re.sub(r'[\\/*?:"<>|]', "_", str(name))

# 转置
df = df.T

print(df.head())





for col in df.columns:
    col_df = df[[col]]

    # 👉 第二行作为文件名（核心）
    second_value = df[col].iloc[0]

    file_name = clean_filename(second_value) + ".pkl"
    file_path = os.path.join(output_dir, file_name)

    col_df.to_pickle(file_path)

print("完成")
```
